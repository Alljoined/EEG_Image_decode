```python
import matplotlib.pyplot as plt
import os
from PIL import Image

obj_paths=[
    "00001_aircraft_carrier/aircraft_carrier_06s.jpg",
    "00024_bug/bug_08s.jpg",
    "00033_cat/cat_01b.jpg"
]

original= os.path.join("/srv/eeg_reconstruction/shared/things_eeg_2/y63gw/osfstorage/test_images/",obj_paths[2])
generated_dir = "/srv/eeg_reconstruction/jonathan/EEG_Image_decode/Generation/Generation/generated_imgs/sub-01/cat"
```

```python
original_image = Image.open(original)
generated_images = [Image.open(os.path.join(generated_dir, f)) for f in os.listdir(generated_dir)]

fig, axes = plt.subplots(1, len(generated_images) + 1, figsize=(15,5))
# Plot original image
axes[0].imshow(original_image)
axes[0].set_title("Original")
axes[0].axis("off")

# Plot generated image
for i, img in enumerate(generated_images):
    axes[i+1].imshow(img)
    axes[i+1].set_title(f"Gen {i+1}")
    axes[i+1].axis("off")

plt.tight_layout()
plt.show()
```
