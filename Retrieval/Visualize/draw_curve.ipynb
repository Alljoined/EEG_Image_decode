```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths for the CSV files
file_paths = ['/home/ldy/Workspace/BrainAligning/outputs/EEGConformer_Encoder_cross_sub.csv', 
              '/home/ldy/Workspace/BrainAligning/outputs/EEGITNet_Encoder_cross_sub.csv',
              '/home/ldy/Workspace/BrainAligning/outputs/EEGNetv4_Encoder_cross_sub.csv']

# Names for each method, for labeling in the plot
method_names = ['Method 1', 'Method 2', 'Method 3']

# Prepare the plot
plt.figure(figsize=(10, 6))

# Loop over each file
for file_path, method_name in zip(file_paths, method_names):
    # Load the CSV file
    data = pd.read_csv(file_path)

    # Extract the necessary columns (assuming 'epoch' and 'test_accuracy' columns)
    epochs = data['epoch']
    test_accuracy = data['test_accuracy']

    # Plot the data
    plt.plot(epochs, test_accuracy, label=method_name, marker='o')

# Adding title, labels, grid, and legend
plt.title('Comparison of Test Set Accuracy Across Different Methods')
plt.xlabel('Epochs')
plt.ylabel('Test Set Accuracy')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
