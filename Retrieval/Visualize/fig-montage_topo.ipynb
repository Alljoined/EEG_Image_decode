```python
import mne
import numpy as np
import matplotlib.pyplot as plt

electrode_regions = {
    'frontal': ['Fp1', 'Fp2', 'AF7', 'AF3', 'AFz', 'AF4', 'AF8', 'F7', 'F5', 'F3', 'F1', 'F2', 'F4', 'F6', 'F8'],
    'central': ['FC5', 'FC3', 'FC1', 'FCz', 'FC2', 'FC4', 'FC6', 'C5', 'C3', 'C1', 'Cz', 'C2', 'C4', 'C6'],
    'parietal': ['CP5', 'CP3', 'CP1', 'CPz', 'CP2', 'CP4', 'CP6', 'P7', 'P5', 'P3', 'P1', 'Pz', 'P2', 'P4', 'P6', 'P8'],
    'temporal': ['FT9', 'FT7', 'T7', 'TP9', 'TP7', 'TP8', 'TP10', 'T8', 'FT8', 'FT10'],
    'occipital': ['PO7', 'PO3', 'POz', 'PO4', 'PO8', 'O1', 'Oz', 'O2']
}

montage = mne.channels.make_standard_montage('standard_1020')

fig, ax = plt.subplots()

pos = montage.get_positions()['ch_pos']

highlighted_electrodes = {ch: region for region, chs in electrode_regions.items() for ch in chs}
colors = {
    'frontal': 'red',
    'central': 'blue',
    'parietal': 'green',
    'temporal': 'yellow',
    'occipital': 'purple'
}

for ch, position in pos.items():
    if ch in highlighted_electrodes:
        ax.plot(position[0], position[1], 'o', color=colors[highlighted_electrodes[ch]], label=highlighted_electrodes[ch], markersize=8)
    else:
        ax.plot(position[0], position[1], 'o', color='gray', markersize=5)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys())

ax.set_title('Highlighted EEG Electrode Regions')
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.set_xticks([])
ax.set_yticks([])
ax.set_aspect('equal')

plt.show()
```
